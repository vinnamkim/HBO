```python
from PESSL import PESSL
import functions
import os
import numpy as np
import copy

runs = 20
init_N = 10
num_Iter = 100

result = []

def cos_vec(x1, x2):
    X = x1.reshape([-1])
    Y = x2.reshape([-1])
    
    return np.abs(np.dot(X, Y)) / (np.sqrt(np.dot(X, X)) * np.sqrt(np.dot(Y, Y)))

for i in xrange(runs):
    fun = functions.sinc_simple10()
    R = PESSL(fun, 1, 5, 1, 0.9, 100, ACQ_FUN = 'EI')
    R_random = PESSL(fun, 1, 5, 1, 0.9, 100, ACQ_FUN = 'EI')
    R_random.data = copy.deepcopy(R.data)
    R_random.fitting()
    
    for j in xrange(num_Iter - init_N):
        vtrue = fun.W
        v1 = R.fitted_params['mu']
        v2 = R_random.fitted_params['mu']
        
        print 'RUN : ' + str(i + 1) + ' ' + 'ITER : ' + str(j + 1) + ' PESSL : ' + str(cos_vec(vtrue, v1)) + ' RANDOM : ' + str(cos_vec(vtrue, v2))
        
        R.iterate(10000, 100)
        R_random.iterate_random()

    result.append(R.data['y'])
    
    result = np.concatenate(result, axis = 1)
    
file_path = "./result/brain_hoo_10D_BSLBO"
directory = os.path.dirname(file_path)

if not os.path.exists(directory):
    os.makedirs(directory)
    
np.save(file_path, result)
```
